```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_n0 = []
X_n1=[]
for n0 in range(1,1000,1):
  for n1 in range(1,1000,1):
    C_n0=(1/n0)/((1/n0)+(1/n1))
    C_n1=(1/n1)/((1/n0)+(1/n1))
    X_n0.append(C_n0)
    X_n1.append(C_n1)
X_n0=np.reshape(X_n0, (-1,999))
X_n1=np.reshape(X_n1, (-1,999))

```

```python
x = np.arange(1,1000,1)
y = np.arange(1,1000,1)
  
X, Y = np.meshgrid(x, y)
#Z_w0 = lossM_w0
#Z_w1 = lossM_w1

 # plotting for w0 initialized

fig = plt.figure(figsize=(20, 8))
ax = fig.add_subplot(1, 2 ,1, projection='3d')
ax.plot_wireframe(X, Y, X_n0, color ='green')
ax.set_title('W0');

# plotting for w1 initialized
ax = fig.add_subplot(1, 2 ,2, projection='3d')
ax.plot_wireframe(X, Y, X_n1, color ='green')
ax.set_title('W1');
```

```python
# plot on 2 dims:
# w0 is in size of 500 , w1 is running:
x_vec= np.arange(1,1000,1)
fig = plt.figure(figsize=(8, 5))
plt.plot(x_vec, X_n1[500, : ])
plt.xlabel("n1")
plt.ylabel("loss")
plt.title("n0=500 n1 run, loss of reset layer 2")
```
